```python
# Load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Since we are using the jupyter notebook, we are going to run a few extra lines for figure to show nicely on screen.

%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basic Line Plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals))
```

```python
# Axis Labels
# The xlabel() and ylabel() functions allow you to set a title for each of the axes
plt.plot(xvals,np.sin(xvals))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
# Multiple plots
plt.plot(xvals,np.sin(xvals))
plt.plot(xvals,np.cos(xvals))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
# Title for the plot using the title function.
plt.plot(xvals,np.sin(xvals))
plt.plot(xvals,np.cos(xvals))
plt.title("Sin/Cos Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# Legend
plt.plot(xvals, np.sin(xvals), label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
# Color the line
plt.plot(xvals, np.sin(xvals), label=r"$y = \sin(x)$", color='red')
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$", color='green')
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
# Color the line - you can give the color names in single letter as well, as below
# Legend
plt.plot(xvals, np.sin(xvals), label=r"$y = \sin(x)$", color='r')
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$", color='g')
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
# Line stype - you can style the plot lines using the linestyle keyword.

plt.plot(xvals, np.sin(xvals), label=r"$y = \sin(x)$", color='r', linestyle='--')
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$", color='g')
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
# Short Format Code 'r--' where r represents color and -- represents linestyle.
plt.plot(xvals, np.sin(xvals), 'r--',label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$", color='g')
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
# Axis limits. You can use the ylim, xlim methods where 1st argument is minimum and 2nd argument is maximum.
plt.plot(xvals, np.sin(xvals), 'r--',label=r"$y = \sin(x)$")
plt.plot(xvals, np.cos(xvals), label=r"$y = \cos(x)$", color='g')
plt.ylim(0,1)
plt.xlim(0,1)
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```
